6.Create two 3*3 matrices with values as first 9 multiples of 3 and 5 respectively. Create a new matrix with output values as sum of the two previous matrices

```python
import numpy as np
list1=[]
for i in range(1,10):
    x=3*i
    list1.append(x)
matrix1 = np.array(list1).reshape(3,3)
list2=[]
for i in range(1,10):
    x=5*i
    list2.append(x)
matrix2 = np.array(list2).reshape(3,3)
res=matrix1+matrix2
res
```

14.Find % change in revenue for each company from 2007 to 2008 [use revenue file from DataSet-2]

16.Find cumulative revenue for 2021 and 2022 [use ytd file from DataSet-2]

```python
ytd=pd.read_csv("ytd.csv")
ytd['bookingdate'] = pd.to_datetime(ytd['bookingdate'], dayfirst=True)
ytd1=ytd.sort_values(by="bookingdate")
ytd1['yyyy'] = pd.to_datetime(ytd1['bookingdate']).dt.year
ytd1['cumsum']=ytd1.groupby('yyyy').cumsum()
ytd1[['bookingdate','yyyy','revenue','cumsum']]
```

17.Plot revenue for each month in 2021 using a suitable plot [use ytd file from DataSet-2]

```python
from matplotlib import pyplot as  plt
ytd1['mm'] = pd.to_datetime(ytd1['bookingdate']).dt.month
ploting_data=ytd1[ytd1['yyyy']==2021]
ploting_data1=ploting_data.groupby('mm')['revenue'].sum()
datapolt=ploting_data1.to_frame(name='sum of revenue in 2021')
datapolt.reset_index()
datapolt.plot(color='green',kind="bar")
```
